```python
import requests

S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

lat = 30.2489#centre of hangzhou
lon = 120.2052
#minlat=117.6251
#maxlat=121.4291
#minlog=31.2809
#maxlog=28.4566
COORDS = ['%f|%f'%(lat,lon)]
PARAMS = {
            'action':"query",
            'list':"geosearch",
            'gscoord': COORDS,
            'gsradius':10000,
            'gslimit':1000,
            'gsprimary': "all",
            'format':"json"
        }

R = S.get(url=URL, params=PARAMS)
DATA = R.json()

PLACES = DATA['query']['geosearch']
for place in PLACES:
    print(place)
    #place['title']
```

```python
len(PLACES)
```

```python
number_elements=len(PLACES)
for i in range(number_elements):
    PLACES[i]['icon']='info-sign'

print (PLACES)
```

```python
d={}
for i in PLACES:
    d[i['title']]=i
print (d)
```
